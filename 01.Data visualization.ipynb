```r
movies <- read.csv("moviesData.csv")
movies
```

```r
ggplot(data = movies, aes(x = critics_score, y = audience_score)) +
  geom_point()
```

```r
ggplot(data = movies, aes(x = critics_score, y = audience_score, color=genre)) +
  geom_point()
```

```r
ggplot(data = movies, aes(x = critics_score, y = audience_score, color=genre)) +
  geom_point()+str(movies$mpaa_rating)
```

```r
ggplot(data = movies, aes(x = critics_score, y = audience_score, color=genre)) +
  geom_point()+str(movies$mpaa_rating)
levels(movies$mpaa_rating)
```

```r
ggplot(data = movies, aes(x = mpaa_rating)) +
  geom_bar()
```

```r
ggplot(data = movies, aes(x = mpaa_rating)) +
  geom_bar()+labs(x="Movies", y="Values", title="Rating")
```

```r
ggplot(data = movies, aes(x = mpaa_rating, fill=genre)) +
  geom_bar()+labs(x="Movies", y="Values", title="Rating")
```

```r
ggplot(data = movies, aes(x = runtime)) +
  geom_histogram()+labs(x="Movies", y="Values", title="Rating")
```
